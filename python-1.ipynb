
# Keywords, Identifiers, Variables and Data types

Keywords: Reserved words that have a specific meaning, 35 keywords

Identifiers: User define names to represent a variable, function, class or any object

Naming convensions for identifiers:
- lowercase (a to z) and uppercase (A to Z) letters
- digits (0 to 9)
- Can have underscore (_)
- Can not use any other special characters
- Can not start with digits
- Should not use keywords as identifiers

Variables: Represents an entity whose value can change


```python
import keyword 
print(keyword.kwlist)
print(len(keyword.kwlist))
```

```python
# Type casting -> changing from one type to another
# 1. Implicit (automatically)
# 2. Explicit (manually)

n1 = 5
n2 = 6.5

print(f"n1 = {n1} of type: {type(n1)}")
print(f"n2 = {n2} of type: {type(n2)}")
```

```python
sum1 = n1 + n2  # implicit

print(f"sum1 = {sum1} of type: {type(sum1)}")
```

```python
sum2 = n1 + int(n2)  # explicit

print(f"sum2 = {sum2} of type: {type(sum2)}")
```

# User Input and Exceptions

```python
name = input("Enter your name")
print(f"Your name is {name}")
```

```python
# Take two numbers as input n1 nd n2
# Divide n1 by n2 and display answer

n1 = float(input("Enter first number: "))
n2 = float(input("Enter second number: "))

answer = n1 / n2
# answer = float(n1) / float(n2)

print(f"{n1} / {n2} = {answer}")
```

```python
# Exception handling

try:
    # code
except:
    # code if exception occur
finally:
    # must execute
```

```python
try:
    n1 = float(input("Enter first number: "))
    n2 = float(input("Enter second number: "))

    answer = n1 / n2
    # answer = float(n1) / float(n2)

    print(f"{n1} / {n2} = {answer}")
except ValueError:
    print("Error: Input must be number")
except Exception as e:
    print(e)
```
