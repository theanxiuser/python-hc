````python
# Take two coordinates input from user and calculate distance
# Use get_coordinate() function for user input
# calculate_distance() function for distance calculation
# main() function for main code
# call main() to execulte whole program
````

````python
import math

def get_coordinate():
    points = []
    for i in range(2):
        p = input(f"Enter x, y for point {i+1}").split(",")
        points.append(tuple(p))
    return points[0], points[1]

def calculate_distance(p1, p2):
    return math.sqrt(
        pow(float(p2[0]) - float(p1[0]), 2) + 
        pow(float(p2[1]) - float(p1[1]), 2)
    )

def main():
    p1, p2 = get_coordinate()
    distance = calculate_distance(p1, p2)
    
    print(f"The distance between points {p1} and {p2} is {distance}.")

main()
````
