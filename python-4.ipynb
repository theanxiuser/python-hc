# Function

## Default arguments, *args, & **kwargs

- Positional argument (*args)
- Keyword argument (**kwargs)

```python
# default argument

def multiply(a, b=1, c=1):
    return a*b*c
```

```python
multiply(4, 5)
```

```python
multiply(4, 5, 6)
```

```python
multiply(4)
```

```python
# positional argument

def info(name, age, *args):
    print(f"Name: {name}")
    print(f"Age: {age}")
```

```python
info("Ram", 20)
```

```python
def info(name, age, *args):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(args)
```

```python
info("Ram", 20, "Kathmandu", "Civil")
```

```python
# Keyword arguments

def info(name, age, *args, **kwargs):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(args)
    print(kwargs)
```

```python
info("Ram", 20, "Kathmandu", "Civil", roll=1)
```

```python
def description(*args, **kwargs):
    print("positional arguments", args)
    print("keyword arguments", kwargs)
    for i in kwargs:
        print(f"{i} = {kwargs[i]}")
```

```python
description(
    name="Ram",
    semester="First",
    department="Computer"
)
```

# File handling

1. Read
2. Write

-> Normal text file, csv file, binary file

## Modes of operation

1. Read only (r)
2. Read and write (r+)
3. Write only (w)
4. Write and read (w+)
5. Append (a)
6. Append and read (a+)

For binary, just add b for example. rb, wb, rb+, wb+, ab, ab+

Steps:
1. Open file
2. Operation 
3. Close file

```python
# open the file

file = open("myfile.txt", "w")

# operation
file.write("This is new file.")

print("Written to file successfully")

# close the file
file.close()
```

```python
file = open("myfile.txt", "r")
data = file.read()
print(data)
file.close()
```

```python
# automatic close file

with open("greeting.py", "r") as f:
    print(f.read())
```

```python
fname = "hello.py"

with open(fname, "r") as f:
    while True:
        data = f.readline()
        if len(data) <= 0:
            break
        print(data, end="")
```

```python
# open a file in w+ mode, 
# write and then read written contents

fname = "file.txt"

with open(fname, "w+") as f:
    f.write("Hello, world\n")
    f.write("Working on file")

    # move filepointer to first position
    f.seek(0)

    data = f.read()
    print("Content of the file:")
    print(data)
```

```python
# binary file
fname = "binary_file"

with open(fname, "wb+") as f:
    f.write(b"Hello, world\n")
    f.write(b"Working on file")

    # move filepointer to first position
    f.seek(0)

    data = f.read()
    print("Content of the file:")
    print(data)
```

```python
# writing into csv file using list

import csv

students = [
    ["Name", "Roll", "Department"],
    ["Ram", 1, "Civil"],
    ["Shyam", 1, "Computer"]
]

with open("student.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(students)
    print("Written successfully.")
```

```python
# reading csv file in lists

with open("student.csv") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
# reading csv file in dict

with open("student.csv") as f:
    reader = csv.DictReader(f)
    # print(reader)
    for row in reader:
        print(f"Name = {row["Name"]}")
        print(f"Roll No = {row["Roll"]}")
        print(f"Department = {row["Department"]}")
        print("==============")
```
