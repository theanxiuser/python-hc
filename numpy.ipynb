````python
# Matrix operations

mat1 = np.array((1, 3, 5, 7)).reshape(2, 2)
mat2 = np.array((2, 4, 6, 8)).reshape(2, 2)

print("Matrix 1: \n", mat1)
print("Matrix 2: \n", mat2)
````

````python
# dot product

np.dot(mat1, mat2)
````

````python
vect1 = np.array((1, 2))
vect2 = np.array((3, 4))

np.dot(vect1, vect2)
````

````python
# matrix multiplication

np.matmul(mat1, mat2)
````

````python
# element wise multiply

mat1 * mat2
````

````python
# Matrix transpose

print("Matrix")
print(mat1)
print("Its transpose")
print(mat1.T)
````

````python
# Inverse

print(mat1)
print("Its inverse")
print(np.linalg.inv(mat1))
````

````python
# Determinant

print(mat1)
print("Its determinant")
print(np.linalg.det(mat1))
````

````python
# Eigen values and eigen vectors

print(mat1)
print("Its eigenvalues and eigenvectors")
eigenvalues, eigenvectors = np.linalg.eig(mat1)
print(eigenvalues)
print(eigenvectors)

# print(np.linalg.eig(mat2))
````

### Q. Solving linear equation
```
x+y+z=6
2x+2y+3z=14
3x+4y+z=10
```

````python
A = np.array(((1, 1, 1), (2, 2, 3), (3, 4, 1)))
B = np.array((6, 14, 10))

X = np.linalg.solve(A, B)

print("A\n", A)
print("B\n", B)
print("X\n", X)
````
