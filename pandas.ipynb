```python
# import pandas

import pandas as pd
```

```python
# Series

srs = pd.Series(["Ram", "Shyam", "Gopal"], name="Names")
print(srs)
```

```python
temp = pd.Series([10, 13, 24, 25], name="Temperature")
print("Temperature\n", temp)
```

```python
# DataFrame

data = {
    "Name": ["Ram", "Shyam", "Gopal", "Sita"],
    "Age": [20, 22, 23, 21],
    "Department": ["Civil", "BCE", "BEE", "BCE"]
}

students = pd.DataFrame(data)
students
```

```python
filtered_data = students[students["Age"] > 21]
filtered_data
```

```python
# save students to csv

students.to_csv("student_data.csv", index=False)
```

```python
# save students to excel
# Need openpyxl module
# !pip install openpyxl

students.to_excel("student_data.xlsx", index=False)
```
