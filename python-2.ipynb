# Loops in python

1. for loop
2. while loop

<pre>
initialize variable (optional)

while condition:
    body of loop
    update variable (optional)
    
</pre>

<pre>
for iterations:
    body of the loop
</pre>

Use case for while loop:
- No of iterations is unknown

Use case for for loop:
- If we know the number of iterations
- Loop through objects/data structures

break and continue in loop

````python
# initialize variable (optional)

# while condition:
#     body of loop
#     update variable (optional)

i = 0

while i < 100:
    print(f"{i + 1}. Hello world!", end=" ")
    i += 1
````

````python
for i in range(0, 100, 10):
    print("Hello world")
````

````python
for i in "Hello world":
    print(i)
````

````python
# 1 * 1 = 1
# 1 * 2 = 2

# 10 * 10 = 100

for i in range(1, 11):
    print("===============")
    for j in range(1, 11):
        print(f"{i} * {j} = {i*j}")
````

````python
i = 1

while i < 10:
    if i == 5:
        i += 1
        continue
    if i == 8:
        break
    print(f"{i}")
    i += 1
    
print("i = ", i)
````

### Problem: Temperature Monitoring System in a Factory

Scenario:
You are designing a temperature monitoring system for a factory. The system reads the temperature of a machine every second and takes action based on the readings.

Requirements:

Read the temperature of the machine 10 times (simulate readings with random values between 20°C and 120°C).

For each temperature reading, apply the following control logic:

- If the temperature is below 50°C, print: "Temperature is safe."
- If the temperature is 50°C to 80°C, print: "Temperature is moderate. Monitor closely."
- If the temperature is above 80°C, print: "Temperature is high! Activate cooling system!"

Keep track of how many times the temperature went into each range (safe, moderate, high) and print the counts at the end.

````python
import random, time

safe_count = 0
moderate_count = 0
high_count = 0

print("Temperature Monitoring System")

for i in range(10):
    print(f"=== Reading {i+1} ===")

    temp = random.randint(20, 120)

    if temp < 50:
        print(f"Temperature = {temp}\u00B0C.")
        print("Temperature is safe.")
        safe_count += 1
    elif 50 <= temp <= 80:
        print(f"Temperature = {temp}\u00B0C.")
        print("Temperature is moderate. Monitor closely.")
        moderate_count += 1
    else:
        print(f"Temperature = {temp}\u00B0C.")
        print("Temperature is high! Activate cooling system!")
        high_count += 1

    time.sleep(1)

print("=== Statistics ==")
print(f"Safe count = {safe_count}")
print(f"Moderate count = {moderate_count}")
print(f"High count = {high_count}")
````
